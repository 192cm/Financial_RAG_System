```python
import torch
import psutil
import platform
import warnings; warnings.filterwarnings('ignore')

print("=== GPU 정보 ===")
# 1. GPU 사용 가능 여부 (True가 나와야 함)
print(f"GPU 이용 가능: {torch.cuda.is_available()}")

# 2. 현재 설치된 PyTorch가 인식하는 CUDA 버전
print(f"PyTorch의 CUDA 버전: {torch.version.cuda}")

# 3. 내 그래픽카드 이름 확인
if torch.cuda.is_available():
    print(f"사용 중인 GPU 이름: {torch.cuda.get_device_name(0)}")

print("\n=== CPU 정보 ===")
# 4. CPU 모델 및 코어 수 확인
print(f"CPU: {platform.processor()}")
print(f"CPU 코어 수: {psutil.cpu_count(logical=False)} (물리), {psutil.cpu_count(logical=True)} (논리)")

print("\n=== RAM 정보 ===")
# 5. RAM 정보 확인
mem = psutil.virtual_memory()
print(f"전체 RAM: {mem.total / (1024**3):.2f} GB")
```
